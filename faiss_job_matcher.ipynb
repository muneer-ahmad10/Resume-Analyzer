```python
# jobs = [
#     "Python NLP Engineer with experience in transformers and semantic search",
#     "Data Analyst skilled in SQL Tableau and Power BI",
#     "Machine Learning Engineer with PyTorch and MLOps experience",
#     "AI Engineer working with LLMs RAG and Vector Databases",
#     "Backend Developer using Java Spring Boot"
# ]

jobs = [
    {
        "title":"NLP Engineer",
        "company":"ABC Tech",
        "skills":"Python NLP Transformers"
    },

    {
        "title":"Data Analyst",
        "company":"XYZ Analytics",
        "skills":"SQL Tableau Power BI"
    },

    {
        "title":"AI Engineer",
        "company":"FutureAI",
        "skills":"LLM RAG FAISS"
    }
]
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)

```

```python
job_texts = [
    f"{job['title']} at {job['company']}. Required skills: {job['skills']}"
    for job in jobs
]

job_embeddings = model.encode(job_texts)
```

```python
import faiss
import numpy as np

dimension = job_embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(
    np.array(job_embeddings).astype("float32")
)
```

```python
resume = """
Python NLP Engineer
Experience with transformers
semantic search
LLMs
"""
```

```python
resume_embedding = model.encode([resume])
```

```python
texts = []

for job in jobs:
    texts.append(
        job["title"] + " " +
        job["skills"]
    )

job_embeddings = model.encode(texts)
```

```python
# distances, indices = index.search(
#     np.array(resume_embedding).astype("float32"),
#     k=3
# )


distances, indices = index.search(
    resume_embedding.astype("float32"),
    3
)
```

```python
# for idx in indices[0]:
#     print(jobs[idx])

for idx in indices[0]:

    print("Title:", jobs[idx]["title"])
    print("Company:", jobs[idx]["company"])
    print("Skills:", jobs[idx]["skills"])
    print()
```
